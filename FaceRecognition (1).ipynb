```python
import face_recognition
from PIL import Image, ImageOps
import os
import pandas as pd

def preprocess_and_save_patches_pil_grayscale(image_folder, csv_file, target_size=(224, 224), save_dir='/content/drive/MyDrive/kaggle challenge/tprocessed_images'):
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    img_labels = pd.read_csv(csv_file)

    for index, row in img_labels.iterrows():
        image_name = row[1]
        image_path = os.path.join(image_folder, image_name)

        try:
            # Load the image in RGB mode for face_recognition
            img = Image.open(image_path)
            img_array = face_recognition.load_image_file(image_path)
            face_locations = face_recognition.face_locations(img_array)

            # Convert the image to grayscale after face detection
            img = img.convert('L')

            if len(face_locations) == 0:
                # No faces detected, take a central patch of the image
                w, h = img.size
                start_x = max(0, w // 2 - target_size[0] // 2)
                start_y = max(0, h // 2 - target_size[1] // 2)
                end_x = start_x + target_size[0]
                end_y = start_y + target_size[1]
                patch = img.crop((start_x, start_y, end_x, end_y))
            else:
                # Process the first detected face for simplicity
                top, right, bottom, left = face_locations[0]
                face = img.crop((left, top, right, bottom))
                patch = ImageOps.fit(face, target_size, method=Image.ANTIALIAS, centering=(0.5, 0.5))

            # Ensure the patch is in grayscale ('L') mode before saving
            patch = patch.convert('L')  # This should be redundant but is a good practice

            save_path = os.path.join(save_dir, f"processed_{image_name}")
            patch.save(save_path)
            print(f"Processed image saved to {save_path}")

        except Exception as e:
            print(f"An error occurred while processing {image_name}: {e}. Saving a blank image.")
            # Save a blank image if an error occurs
            blank_image = Image.new('L', target_size, (0))  # Create a blank grayscale image
            save_path = os.path.join(save_dir, f"processed_{image_name}")
            blank_image.save(save_path)

# Example usage
image_folder = '/content/drive/MyDrive/train'
csv_file = '/content/drive/MyDrive/kaggle challenge/train.csv'


preprocess_and_save_patches_pil_grayscale(image_folder, csv_file)
```

```python
import torch
import torch.nn as nn
import torchvision.models as models
from torchvision.models import ResNet50_Weights
from torch.utils.data import DataLoader
from torchvision import transforms
import pandas as pd
from torch.utils.data import Dataset, DataLoader
```

```python
class ProcessedFacesDataset(Dataset):
    def __init__(self, img_dir, csv_file, category_file, transform=None):
        self.img_labels = pd.read_csv(csv_file)
        # Load the categories and implicitly use their row index as the class index
        self.categories = pd.read_csv(category_file)['Category'].tolist()
        # Mapping from category names to their indices
        self.category_to_index = {category: i for i, category in enumerate(self.categories)}

        self.img_dir = img_dir
        self.transform = transform

    def __len__(self):
        return len(self.img_labels)

    def __getitem__(self, idx):
        img_path = os.path.join(self.img_dir, f"processed_{self.img_labels.iloc[idx, 0]}.jpg")
        image = Image.open(img_path).convert('RGB')
        if self.transform:
            image = self.transform(image)

        # Map the label name to its index
        label_name = self.img_labels.iloc[idx]['Category']
        label_index = self.category_to_index[label_name]

        return image, label_index
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

dataset = ProcessedFacesDataset(img_dir='/content/drive/MyDrive/kaggle challenge/tprocessed_images',
                                csv_file='/content/drive/MyDrive/kaggle challenge/train.csv',
                                category_file='/content/drive/MyDrive/kaggle challenge/category.csv',
                                transform=transform)
loader = DataLoader(dataset, batch_size=32, shuffle=True, num_workers=4)
```

```python
from torch.utils.data.dataset import random_split

# Assume the base transform is defined
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Initialize your dataset
dataset = ProcessedFacesDataset(img_dir='/content/drive/MyDrive/kaggle challenge/tprocessed_images',
                                csv_file='/content/drive/MyDrive/kaggle challenge/train.csv',
                                category_file='/content/drive/MyDrive/kaggle challenge/category.csv',
                                transform=transform
                                )

# Calculate the sizes for training and validation sets: 80% - 20% split
total_size = len(dataset)
train_size = int(0.8 * total_size)
valid_size = total_size - train_size

# Split the dataset
train_dataset, valid_dataset = random_split(dataset, [train_size, valid_size])
```

```python
batch_size = 32

# DataLoader for the training dataset
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=4)

# DataLoader for the validation dataset
valid_loader = DataLoader(valid_dataset, batch_size=batch_size, shuffle=False, num_workers=4)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
resnet50.to(device)

# Change from Adam to SGD
optimizer = torch.optim.SGD(resnet50.parameters(), lr=0.01, momentum=0.9)

criterion = nn.CrossEntropyLoss()
num_epochs = 15
for epoch in range(num_epochs):
    # Training phase
    resnet50.train()
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = resnet50(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

    # Validation phase
    resnet50.eval()
    correct_valid = 0
    total_valid = 0
    with torch.no_grad():
        for images, labels in valid_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = resnet50(images)
            _, predicted = torch.max(outputs.data, 1)
            total_valid += labels.size(0)
            correct_valid += (predicted == labels).sum().item()

    valid_accuracy = 100 * correct_valid / total_valid
    print(f'Epoch {epoch+1}, Validation Accuracy: {valid_accuracy}%')
```

```python
import torch
from torchvision import transforms
from torch.utils.data import DataLoader, Dataset
import pandas as pd
import os
from PIL import Image


# Ensure the model is loaded with the correct weights
model_path = 'resnet50_feature_extractor.pth'
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = resnet50  # Your model definition here
model.load_state_dict(torch.load(model_path, map_location=device))
model.to(device)
model.eval()

class ProcessedImageDataset(Dataset):
    def __init__(self, directory, transform=None):
        self.directory = directory
        self.transform = transform
        self.image_filenames = [f for f in os.listdir(directory) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

    def __len__(self):
        return len(self.image_filenames)

    def __getitem__(self, idx):
        img_name = self.image_filenames[idx]
        image_path = os.path.join(self.directory, img_name)
        image = Image.open(image_path).convert('RGB')
        if self.transform:
            image = self.transform(image)
        return image, img_name

# Transformation pipeline
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),

])

# Load the mapping from indices to names
category_data = pd.read_csv('/content/drive/MyDrive/kaggle challenge/category.csv')
category_name_mapping = category_data['Category'].to_dict()
def map_prediction_to_name(prediction):
    # Convert prediction index to celebrity name using the loaded mapping
    return category_name_mapping.get(prediction, "Unknown")

# Create your dataset and dataloader
processed_images_dir = '/content/drive/MyDrive/kaggle challenge/test_processed_images'
dataset = ProcessedImageDataset(directory=processed_images_dir, transform=transform)
dataloader = DataLoader(dataset, batch_size=32, shuffle=False)

# Predict and save to CSV
predictions = []
for inputs, image_names in dataloader:
    inputs = inputs.to(device)
    with torch.no_grad():
        outputs = model(inputs)
        _, predicted = torch.max(outputs, 1)
        predicted = predicted.cpu().numpy()
        for img_name, pred in zip(image_names, predicted):
            predictions.append((img_name, map_prediction_to_name(pred)))

# Save predictions to CSV
prediction_file = 'predictionssgdmore15.csv'
with open(prediction_file, 'w') as f:
    f.write('Id,Category\n')
    for img_name, pred_name in predictions:
        f.write(f"{img_name},{pred_name}\n")

print(f"Predictions saved to {prediction_file}")
```
